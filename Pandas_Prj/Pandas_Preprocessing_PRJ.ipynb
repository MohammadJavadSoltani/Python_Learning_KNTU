## 2. Inconsistent Data:

### It's much harder than Nan values. Frist lets have a point view:

```python
for col in df.columns:
    print(f'{col :30s} >>>>      {df[col].unique()}')
```

Replace:

```python
#Geography
df['Geography'].replace('3', df['Geography'].mode()[0], inplace=True)
df['Geography'].replace('4', df['Geography'].mode()[0], inplace=True)
df['Geography'].replace('56', df['Geography'].mode()[0], inplace=True)
df['Geography'].replace('Pooya', df['Geography'].mode()[0], inplace=True)
df['Geography'].replace('10000', df['Geography'].mode()[0], inplace=True)
#Gender
df['Gender'].replace('9', df['Gender'].mode()[0], inplace=True)
df['Gender'].replace('5', df['Gender'].mode()[0], inplace=True)
#Age
df['Age'].replace(-10, df['Age'].mode()[0], inplace=True)
```

```python
#HasCaCard
df['HasCrCard'].replace(-1000, df['HasCrCard'].mode()[0], inplace=True)

#IsActiveMember 
df['IsActiveMember'].replace(-1.0, df['IsActiveMember'].mode()[0], inplace=True)
```

```python
for col in df.columns:
    print(f'{col :30s} >>>>      {df[col].unique()}')
```
